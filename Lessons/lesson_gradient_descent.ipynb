## Specifics  of Gradient Descent
* iterative, first order optimization algorithm, used to find a maximum accuracy or minimum loss of a given function
* iterates over the training dataset while re-adjusting the model
* an optimization technique that can find the minimum of an objective function
* finds the values of a function‚Äôs parameters that minimize the **cost function** as far as possible
* finds the optimal solution by taking a step in the direction of the maximum rate of decrease of the function

**Note**: *By minimizing the cost function we can find the optimal parameters that yield the best model performance.*

## Approach
In the gradient descent algorithm, we start with random model parameters and calculate the error for each learning iteration, keep updating the model parameters to move closer to the values that results in minimum cost.

## Define a Gradient Descent Function 

* Implements a loop that ends when either:
    * the number of iterations exceeds a maximum value
    * the difference in function values between two successive iterations falls below a certain threshold
* The parameters are updated at every iteration according to the gradient of the objective function - Stochastic

### Gradient Descent for Simle Linear Regression (ùë¶ = ùõΩ<sub>0</sub>+ ùõΩ<sub>1</sub>‚àóùë•)
* The function accepts the following parameters:
    * training set: X and Y
    * learning rate: the step size for gradient descent
    * tolerance: the threshold that will stop the process if the difference in function values between two successive iterations falls below this threshold
    * max_iterations: the maximum number of iterations to run
    * initial intercept: ùõΩ<sub>0</sub><br>
    * initial slope: ùõΩ<sub>1</sub><br>
    
**Note**: 
* **Iterations, learning rate, tolerance, initial intercept (ùõΩ<sub>0</sub>) and initial slope (ùõΩ<sub>1</sub>) are are hyperparameters that can be tuned.**
* **Gradient Descent optimizes the model based on hyperparameters and in order to tune its hyperparameters, GridSearch and RandomSearch are used.**

```python
# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simple Linear Regression cost function
def mean_squared_error(Y, Y_pred):
    cost = np.sum((Y - Y_pred)**2) / len(Y)
    
    return cost
```

```python
# Calculates the gradients of a Simple Linear Regression
def gradient_func(X, Y, Y_pred):
    n = float(len(X))
    
    intercept_derivative = -(2/n) * sum(Y - Y_pred)
    slope_derivative = -(2/n) * sum(X * (Y - Y_pred)) 
    
    return intercept_derivative, slope_derivative
```

```python
# Gradient Descent Function for Simple Linear Regression
def gradient_descent_simple_linear_regression(X, Y, learning_rate = 0.0001, tolerance = 1e-6, iterations = 1000, 
                                              initial_intercept = 0.01, initial_slope = 0.01):
    
    # Initializing slope, intercept, learning rate and iterations
    current_intercept = initial_intercept
    current_slope = initial_slope
    iterations = iterations
    learning_rate = learning_rate
    
    costs = []
    slopes = []
    intercepts = []
    previous_cost = None

    #find the optimal parameters
    for i in range(iterations):

        # make predictions with the current intercept and slope
        Y_pred = current_intercept + (current_slope * X)

        # calculate the current cost
        current_cost = mean_squared_error(Y, Y_pred)

        # Check for the stopping condition provided by the tolerance
        if previous_cost:
            if abs(previous_cost-current_cost)<=tolerance:
                break

        previous_cost = current_cost

        costs.append(current_cost)
        intercepts.append(current_intercept)
        slopes.append(current_slope)

        # calculate the gradients
        intercept_derivative, slope_derivative = gradient_func(X, Y, Y_pred)
        
        # update both the intercept and slope
        current_intercept = current_intercept - (learning_rate * intercept_derivative)
        current_slope = current_slope - (learning_rate * slope_derivative)

        # print the parameters for each iteration
        #print(f"Iteration {i+1}: Cost {current_cost}, Intercept {current_intercept}, Slope {current_slope}")

    # Visualize the intercept and cost for all iterations
    plt.figure(figsize = (8,6))
    plt.plot(intercepts, costs)
    plt.scatter(intercepts, costs, marker='o', color='red')
    plt.title("Cost vs Intercepts")
    plt.ylabel("Cost")
    plt.xlabel("Intercept")
    plt.show()

    # Visualize the slope and cost for all iterations
    plt.figure(figsize = (8,6))
    plt.plot(slopes, costs)
    plt.scatter(slopes, costs, marker='o', color='red')
    plt.title("Cost vs Slopes")
    plt.ylabel("Cost")
    plt.xlabel("Slope")
    plt.show()
    
    return current_intercept, current_slope
```

### Example: Generic Data

```python
# Data
X = np.array([32.50234527, 53.42680403, 61.53035803, 47.47563963, 59.81320787,
              55.14218841, 52.21179669, 39.29956669, 48.10504169, 52.55001444,
              45.41973014, 54.35163488, 44.1640495 , 58.16847072, 56.72720806,
              48.95588857, 44.68719623, 60.29732685, 45.61864377, 38.81681754])
Y = np.array([31.70700585, 68.77759598, 62.5623823 , 71.54663223, 87.23092513,
              78.21151827, 79.64197305, 59.17148932, 75.3312423 , 71.30087989,
              55.16567715, 82.47884676, 62.00892325, 75.39287043, 81.43619216,
              60.72360244, 82.89250373, 97.37989686, 48.84715332, 56.87721319])

estimated_intercept, estimated_slope  = gradient_descent_simple_linear_regression(X, Y, iterations=2000)

print(f"Estimated Intercept: {estimated_intercept}\nEstimated Slope: {estimated_slope}")

# make predictions by using estimated parameters
Y_pred = estimated_intercept + estimated_slope*X

# plot the regression line
plt.figure(figsize = (8,6))
plt.scatter(X, Y, marker='o', color='red')
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='blue',markerfacecolor='red', markersize=10,linestyle='dashed')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

The example data we‚Äôre going to use for Gradient Descent Optimization is on California Housing Prices.

**Objective**:
* Compare Gradient Descent Optimization and Linear Regression Prediction with **sklearn.linear_model.LinearRegression**.

### Step 1. Load Data

```python
import pandas as pd
df = pd.read_csv('data\\housing.csv')
df.head(10)
```

### Step 2. Prepare the Data - X and Y Sets

```python
#Y = np.array(df.iloc[:, df.columns.get_loc('price')].values)
#X = np.array(df.iloc[:, df.columns.get_loc('bedrooms')].values)

Y = df.iloc[:, df.columns.get_loc('price')].values
X = df.iloc[:, df.columns.get_loc('bathrms')].values

Y_df = df.iloc[:, df.columns.get_loc('price')]
X_df = df.loc[:, df.columns == 'bathrms']
```

### Step 3. Run the Gradient Descent with a Small Number of Iterations

```python
n_iterations = 2000
```

```python
estimated_intercept, estimated_slope  = gradient_descent_simple_linear_regression(X, Y, iterations=n_iterations)

print(f"Estimated Intercept: {estimated_intercept}\nEstimated Slope: {estimated_slope}")

# make predictions by using estimated parameters
Y_pred = estimated_intercept + estimated_slope*X

# plot the regression line
plt.figure(figsize = (8,6))
plt.scatter(X, Y, marker='o', color='red')
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='blue',markerfacecolor='red', markersize=10,linestyle='dashed')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
from sklearn import metrics
gr_desc_score = metrics.r2_score(Y, Y_pred)
print(gr_desc_score)
```

### Step 4. Predict Y with LinearRegression and Compare with Gradient Descent

```python
lr = LinearRegression()
lr.fit(X_df, Y_df)
```

```python
Y_pred = lr.predict(X_df)
```

```python
lr_score = metrics.r2_score(Y_df, Y_pred)
print(lr_score)
```

```python
print(gr_desc_score)
print(lr_score)
```
